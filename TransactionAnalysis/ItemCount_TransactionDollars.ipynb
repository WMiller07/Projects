```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt_df = pd.read_csv('./ItemCount_TransactionDollars.csv')
```

```python
plt_df.head(10)
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(x=plt_df['count_Items'], y=plt_df['avg_transaction'], ax=ax)
ax.set_xlim([0,50])
ax.set_ylim([0,300])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(x=plt_df['count_Items'], y=plt_df['count_transactions'], ax=ax)
ax.set_xlim([0,25])
plt.show()
```

```python
filt_itemcount = plt_df['count_Items'] < 7
plt_df[filt_itemcount]['count_transactions'].sum()/plt_df['count_transactions'].sum()
```

```python
plt_df['pct_transactions'] = plt_df['count_transactions']/plt_df['count_transactions'].sum()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(x=plt_df['count_Items'], y=plt_df['pct_transactions'].cumsum(), ax=ax)
ax.set_xlim([1,25])
ax.set_xticks(np.arange(1, 26, 1.0))
plt.show()
```
