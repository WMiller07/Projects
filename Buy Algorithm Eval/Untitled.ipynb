```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_r40 = pd.read_csv('./data/R40_SuggestedOffers_ActualOffers_Comp.csv', header=0)
df_r42 = pd.read_csv('./data/R42_SuggestedOffers_ActualOffers_Comp.csv', header=0)
```

```python
df_r42.info()
```

```python
sns.set_style('whitegrid')
```

```python
filt_r40qtypurch = df_r40['count_BuyItems'] >= 10
filt_r42qtypurch = df_r42['count_BuyItems'] >= 10
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(x='count_BuyItems', y='avg_ItemOffer', data=df_r40[filt_r40qtypurch], ax=ax, alpha = 0.5)
ax.set_xlim([0,200])
ax.set_ylim([0,10])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(x='count_BuyItems', y='Chain_SuggestedOffer', data=df_r40[filt_r40qtypurch], ax=ax, alpha = 0.5)
ax.set_xlim([0,200])
ax.set_ylim([0,10])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(x='count_BuyItems', y='chain_suggestedOffer', data=df_r42[filt_r42qtypurch], ax=ax, alpha = 0.5)
ax.set_xlim([0,200])
ax.set_ylim([0,10])
plt.show()
```
