```python
import datetime
from dateutil.relativedelta import relativedelta
#from datetime import timedelta
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mtick
from matplotlib.ticker import FormatStrFormatter
from matplotlib.lines import Line2D
import matplotlib.animation as animation

import seaborn as sns
```

```python
df_comp = pd.read_csv('./BuyOffersAlgComp.csv', header=0, index_col=0)
df_comp.dropna(inplace=True)
```

```python
sns.set_style('whitegrid')
```

```python
def barchart_GradeDist(df, str_SOLabel, pal='Reds_r', **kwargs):
    filt_Comp = df['CatalogBinding'] == 'General'
    plt_df = df_comp[filt_Comp].groupby(str_SOLabel + 'Pct')[['Quantity', str_SOLabel]].sum()

    fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(24,6))
    
    sns.barplot(plt_df.index, plt_df[str_SOLabel].values, order=plt_df.index.sort_values(ascending=False), palette=pal, ax=ax1)
    sns.barplot(plt_df.index, plt_df['Quantity'].values, order=plt_df.index.sort_values(ascending=False), palette=pal, ax=ax2)
    sns.barplot(plt_df.index, (plt_df[str_SOLabel].values / plt_df['Quantity'].values), order=plt_df.index.sort_values(ascending=False), palette=pal, ax=ax3)
    
    ax1.set_ylim(0, 150000)
    ax1.set_xlabel('Suggested Offer %')
    ax1.set_ylabel('Total Cost')
    ax1.set_xticklabels(['{:.0%}'.format(float(x.get_text())) for x in ax1.get_xticklabels()]) 
    ax1.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}')) 
   
    ax2.set_ylim(0, 150000)
    ax2.set_xlabel('Suggested Offer %')
    ax2.set_ylabel('Total Quantity')
    ax2.set_xticklabels(['{:.0%}'.format(float(x.get_text())) for x in ax2.get_xticklabels()]) 
    ax2.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}')) 
    
    ax3.set_ylim(0, 8)
    ax3.set_xlabel('Suggested Offer %')
    ax3.set_ylabel('Average Offer Amount')
    ax3.set_xticklabels(['{:.0%}'.format(float(x.get_text())) for x in ax3.get_xticklabels()]) 
    ax3.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.2f}')) 
    
    plt.suptitle('Suggested Offer Algorithm Grade Distribution')
    plt.show()
    return
```

```python
barchart_GradeDist(df_comp, 'r40_SuggestedOffer', pal='Blues_r')
```

```python
barchart_GradeDist(df_comp, 'r41_SuggestedOffer', pal='Greens_r')
```
