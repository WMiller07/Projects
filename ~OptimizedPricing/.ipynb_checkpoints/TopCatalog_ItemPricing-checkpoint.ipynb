```python
def createPriceHistogram(arr, boolDensity=False, boolArea=False, maxPrice = 50, increments = 0.5, **kwargs):
    """
    Returns: (a, b) where "a" and "b" are each 1-d NumPy arrays, together describing a histogram encompassing all pricepoints for a 
        catalog title below a certain threshold and increments, weighted by another array.
        "a" -- describes the bins boundaries of the histogram. 
        "b" -- describes the values within each bin.
    
    Keyword arguments:
        arr -- a 1-d array to be converted into a histogram
        boolDensity -- boolean argument, if true will return a density distrubution with a total area of 1.
        boolArea -- boolean argument, if true will use the input array as weights on itself, making each element of the histogram equal
            to [count * thingCounted]. Is only valid for histograms of numerical data.
        maxPrice -- sets the upper bound of the histogram
        increments -- sets the bin increments of the histogram such that the number of bins will be equal to [maxPrice / increments]
    """
    n_bins = int(maxPrice / increments)
    
    if boolArea == False:
        arr_dist, arr_bins = np.histogram(arr, bins = n_bins, range=[0, maxPrice], density=boolDensity) 
    else:
        arr_dist, arr_bins = np.histogram(arr, bins = n_bins, range=[0, maxPrice], weights=arr, density=boolDensity) 
        
    if boolDensity == True:
        #numpy histogram density only sums to 1 if increments are 1, necessitating conversion for smaller values
        arr_dist = arr_dist * increments
    return np.array(arr_dist), np.array(arr_bins)


def createWeightedPriceHistogram(arr, w, maxPrice = 50, increments = 0.5, **kwargs):
    """
    Returns: (a, b) where "a" and "b" are each 1-d NumPy arrays, together describing a histogram encompassing all pricepoints for a 
        catalog title below a certain threshold and increments, weighted by another array.
        "a" -- describes the bins boundaries of the histogram. 
        "b" -- describes the values within each bin.
    
    Keyword arguments:
        arr -- a 1-d array to be converted into a histogram.
        w -- a 1-d array used as weights on arr
        maxPrice -- sets the upper bound of the histogram
        increments -- sets the bin increments of the histogram such that the number of bins will be equal to [maxPrice / increments]
    """
    n_bins = int(maxPrice / increments)
    arr_dist, arr_bins = np.histogram(arr, bins = n_bins, weights=w, range=[0, maxPrice]) 

    return np.array(arr_dist), np.array(arr_bins)


def commonPricePoints(arr, densityThreshold = 0.1, **kwargs):
    """
    Returns: (a, b) where "a" and "b" are each 1-d NumPy arrays. "a" is an array of the bins of the histogram of all price points.
        "a" -- an array of boolean values with "True" describing which values of "b" transcends a commonality threshold 
        "b" -- describes the values within each bin.
    Keyword arguments:
        arr -- a 1-d array to be converted into a histogram.
        densityThreshold -- the threshold above which a price point will be considered "common"
    """
    arr_density, arr_bins = createPriceHistogram(arr, boolDensity = True)
    boolarr_commonBins = [d >= densityThreshold for d in arr_density]
    
    return np.array(boolarr_commonBins), np.array(arr_bins)

def amtSoldAtPricePoint(df):
    filt_SalesAtPricePoint = df['OriginalPrice'] == df['RegisterPrice']
    arr_OriginalPriceSales, _ = createPriceHistogram(df[filt_SalesAtPricePoint]['OriginalPrice'], boolArea=True)
    arr_OriginalPriceSellThrough = arr_OriginalPriceSales 
    return np.array(arr_OriginalPriceSellThrough)

def qtyPricedAtPricePoint(df):
    arr_qtyPricedOriginalPrice, _ = createPriceHistogram(df['OriginalPrice'])
    return np.array(arr_qtyPricedOriginalPrice)

def qtySoldAtPricePoint(df):
    filt_SalesAtPricePoint = df['OriginalPrice'] == df['RegisterPrice']
    arr_qtySoldOriginalPrice, _ = createPriceHistogram(df[filt_SalesAtPricePoint]['OriginalPrice']) 
    return np.array(arr_qtySoldOriginalPrice)

def qtySoldInDaysAtPricePoint(df, daysToSell):
    filt_SalesAtPricePoint = df['OriginalPrice'] == df['RegisterPrice']
    filt_DaysToSellThreshold = df['DaysToSell'] <= daysToSell
    arr_qtySoldInDays, _ = createPriceHistogram(df[filt_SalesAtPricePoint & filt_DaysToSellThreshold]['OriginalPrice']) 
    return np.array(arr_qtySoldInDays)

# def ageAtPricePoint(df):
#     filt_SalesAtPricePoint = df['OriginalPrice'] == df['RegisterPrice']
#     arr_ageAtPricePoint, _ = createWeightedPriceHistogram(df[filt_SalesAtPricePoint]['OriginalPrice'], 
#                                                           df[filt_SalesAtPricePoint]['DaysToSell'])
#     return np.array(arr_ageAtPricePoint)

def optimalPricePoint(df, daysOnShelfThreshold = 90, densityThreshold = 0.1, verbose = 0, **kwargs):
    c, b = commonPricePoints(df['OriginalPrice'], densityThreshold = densityThreshold)
    #d = ageAtPricePoint(df)
    qds = qtySoldInDaysAtPricePoint(df, daysOnShelfThreshold)
    qs = qtySoldAtPricePoint(df)
    qp = qtyPricedAtPricePoint(df)
    a = amtSoldAtPricePoint(df)
    if verbose != 0:
        df_summary = pd.DataFrame({'priceBin': b[0:-1], 'qtyPriced': qp, 'qtySold': qs, 'amtSold': a, 'qtySoldIn90Days': qds, 'isCommon': c})
        if verbose == 1:
            print(df_summary[df_summary['isCommon'] == True])
        else:
            print(df_summary)
    # return avgPricePoint at the maximum value of PricePoint_SellThrough_AtAgeThreshold * PricePoint
    return round(np.divide(a, qs, where = qs!=0)[(np.divide(qds, qp, where = qp!=0) * np.divide(a, qs, where = qs!=0) * c).argmax()], 2)
```

```python
dict_c = {}
for c in df['CatalogId'].unique():
    filter_CatalogID = df['CatalogId'] == c
    df_t = df[filter_CatalogID].copy()
    optimalPrice = optimalPricePoint(df_t)
    sellThrough = df_t['qtySold'] / df_t['qtyPriced']
    sellThrough90Day = df_t['qtySoldIn90Days'] / df_t['qtyPriced']
    itemCount = len(df_t)
    dict_c[c] = {'optimalPrice': optimalPrice, 'itemCount': itemCount, 'sellThrough' = sellThrough, sellThrough90Day = }

df_optimalprices = pd.DataFrame.from_dict(dict_c, orient='index')
```
